```python
import pandas as pd
import numpy as np
```

```python
#load ontario-2015 csv
df1 = pd.read_csv(r"C:\Users\HP\Downloads\ONTARIO-2015.csv",encoding='ISO-8859-1')
df1
```

```python
df1.columns
```

```python
#load ontario-2016 csv
df2 = pd.read_csv(r"C:\Users\HP\Downloads\ONTARIO-2016.csv",encoding='ISO-8859-1')
df2
```

```python
#concat ontario 2015 & 2016
df = pd.concat([df1,df2],ignore_index=True)
df
```

```python
#missing values
df.isnull().sum()
```

```python
#remove duplicates
df = df.drop_duplicates()
df.shape
```

```python
# these columns have more 700 nan rows so deleting these columns
df = df.drop(columns=['Max Temp Flag','Min Temp Flag','Mean Temp Flag','Heat Deg Days Flag','Cool Deg Days Flag','Total Rain Flag','Total Snow Flag',
        'Total Precip Flag','Snow on Grnd Flag','Dir of Max Gust (10s deg)','Dir of Max Gust Flag','Spd of Max Gust (km/h)','Spd of Max Gust Flag'])
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# List of columns you want to impute with the mean
columns_to_impute = ['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)', 
                     'Heat Deg Days (°C)', 'Cool Deg Days (°C)', 
                     'Total Rain (mm)', 'Total Snow (cm)', 'Total Precip (mm)','Snow on Grnd (cm)']

# Loop through each column and fill missing values with the mean
for col in columns_to_impute:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Data Quality'].value_counts()
```

```python

# Replace NaN with 'Data Not Recorded' in the 'Data Quality' column
df['Data Quality'] = df['Data Quality'].fillna('Data Not Recorded')

# Replace '' with 'Unverified Data' in the 'Data Quality' column
df['Data Quality'] = df['Data Quality'].replace('', 'Unverified Data')
```

```python
df.isnull().sum()
```

```python
df['Data Quality'].value_counts()
```

```python
#datatype
df.dtypes
```

```python
cont_col=cat_col = df.select_dtypes(exclude=['object','datetime64[ns]']).columns
cont_col
```

```python
#check outliers
for column in cont_col:

    print(f"Column : {column}")
    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_threshold = df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = df[column].quantile(0.25) - (1.5 * iqr)
    print(f"UpperThreshold : {upper_threshold} and lowerThreshold : {lower_threshold}")

    print(df[column].describe())
    print('**********************')
```

```python
outlier = df[['Max Temp (°C)', 'Min Temp (°C)',
       'Mean Temp (°C)', 'Heat Deg Days (°C)', 'Cool Deg Days (°C)',
       'Total Rain (mm)', 'Total Snow (cm)', 'Total Precip (mm)',
       'Snow on Grnd (cm)']]
```

```python
#remove outlier
for column in outlier:

    print(f"Column : {column}")

    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_threshold = df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = df[column].quantile(0.25) - (1.5 * iqr)
    print(f"UpperThreshold : {upper_threshold} and lowerThreshold : {lower_threshold}")

    df[column] = df[column].clip(lower_threshold, upper_threshold)   #cliping- removing outliers
    print(df[column].describe())


    print('**********************')
```

```python
df.to_csv("merged_ontario.csv",index=False) # save the merged dataset
```
