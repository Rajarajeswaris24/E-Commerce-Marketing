```python
import pandas as pd

# Assuming `df` is your DataFrame

# Convert the order_item_id and order_id to integers if they are in scientific notation
merged_df['order_id'] = merged_df['order_id'].apply(lambda x: int(x) if pd.notnull(x) else x)
merged_df['order_item_id'] = merged_df['order_item_id'].apply(lambda x: int(x) if pd.notnull(x) else x)

```

```python

# Check for empty strings in each cell of the DataFrame
empty_strings = merged_df.applymap(lambda x: x == ' ')

# Get the column names with empty strings
columns_with_empty_strings = empty_strings.any()

# Print the column names with empty strings
print("Columns with empty strings:")
print(columns_with_empty_strings[columns_with_empty_strings].index.tolist())
```

```python
merged_df['gmv'] = merged_df['gmv'].replace(' ',np.nan) #changing gmv to float
```

```python
merged_df['gmv'].isnull().sum()
```

```python
merged_df['gmv'] = pd.to_numeric(merged_df['gmv'], errors='coerce')

# Fill NaN values with the mean of the 'gmv' column
mean_value = merged_df['gmv'].mean()
merged_df['gmv'] = merged_df['gmv'].fillna(mean_value)

# Convert the 'gmv' column to integers
merged_df['gmv'] = merged_df['gmv'].astype(int)
```

```python
outlier = merged_df[[ 'order_id', 'order_item_id','gmv', 'units', 'sla',
       'product_mrp', 'product_procurement_sla']]
```

```python
for column in outlier:  #checking outliers

    print(f"Column : {column}")
    iqr = merged_df[column].quantile(0.75) - merged_df[column].quantile(0.25)
    upper_threshold = merged_df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = merged_df[column].quantile(0.25) - (1.5 * iqr)
    print(f"UpperThreshold : {upper_threshold} and lowerThreshold : {lower_threshold}")

    print(merged_df[column].describe())
    print('**********************')
```

```python
for column in outlier:

    print(f"Column : {column}")

    iqr = merged_df[column].quantile(0.75) - merged_df[column].quantile(0.25)
    upper_threshold = merged_df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = merged_df[column].quantile(0.25) - (1.5 * iqr)
    print(f"UpperThreshold : {upper_threshold} and lowerThreshold : {lower_threshold}")

    merged_df[column] = merged_df[column].clip(lower_threshold, upper_threshold)   #cliping- removing outliers
    print(merged_df[column].describe())


    print('**********************')
```
